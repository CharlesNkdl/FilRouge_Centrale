```python
from sklearn.preprocessing import LabelEncoder

l_e = LabelEncoder()
df.head()
```

```python
tlt = ["date_urg","date_box","date_prescri_bio","date_prelev_bio","date_sortie","date_deb_pec_med"]
for i in tlt:
    df[i] = pd.to_datetime(df[i], format='%Y-%m-%d %H:%M:%S')

def minutes_since_midnight(time):
    return time.hour * 60 + time.minute

# Fonction pour calculer le delta en minutes
def calculate_delta(start, end):
    
    delta = pd.to_datetime(end) - pd.to_datetime(start)
    if delta.total_seconds() <= 0:
        return np.nan
    return delta.total_seconds() / 60

# Calculs
df['arrivee_minutes'] = df['date_urg'].apply(minutes_since_midnight)
df['delta_arrivee_pec_med'] = df.apply(lambda row: calculate_delta(row['date_urg'], row['date_deb_pec_med']), axis=1)
df['delta_arrivee_box'] = df.apply(lambda row: calculate_delta(row['date_urg'], row['date_box']), axis=1)
df['delta_prescri_bio'] = df.apply(lambda row: calculate_delta(row['date_urg'], row['date_prescri_bio']), axis=1)
df['duree_prelev_bio'] = df.apply(lambda row: calculate_delta(row['date_prescri_bio'], row['date_prelev_bio']), axis=1)
df['delta_arrivee_sortie'] = df.apply(lambda row: calculate_delta(row['date_urg'], row['date_sortie']), axis=1)

df.head()
```

```python
txt = ['have_bio', 'have_radio', 'have_echo', 'have_scan','have_irm', 'have_specia']
for x in txt:
	df[x] = l_e.fit_transform(df[x])
df.head()
```

```python
df = df.drop(["date_urg","date_box","date_prescri_bio","date_prelev_bio","date_sortie","date_deb_pec_med"],axis=1)
```

```python
def round_up_to_nearest_15(minutes):
    return np.ceil(minutes / 15) * 15

x = ["arrivee_minutes","delta_arrivee_pec_med","delta_arrivee_box","delta_prescri_bio","duree_prelev_bio","delta_arrivee_sortie"]
for t in x:
    df[t] = df[t].apply(round_up_to_nearest_15)
df.head()
```

```python
df['fc_ad'] = df['fc_ad'].apply(lambda x: np.nan if not isinstance(x, int) else x)
#df = df.drop(["cd_ad"],axis=1)
```

```python
df["ccmu"] = df["ccmu"].apply(lambda x: np.nan if not isinstance(x, int) else x)
```

```python
from sklearn.impute import SimpleImputer

x = ["fc_ad","tmp_ad","systo","diasto","delta_arrivee_pec_med","delta_arrivee_box","delta_prescri_bio","duree_prelev_bio"]
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

x = df.drop("delta_arrivee_sortie", axis = 1)
y = df["delta_arrivee_sortie"]
# Standardisation des caractéristiques (excluant la cible)
X_scaled = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA

# Choix du nombre de composantes
pca = PCA(n_components=5)  # Exemple : réduire à 10 composantes principales


X_pca = pca.fit_transform(X_scaled)
print(pca.explained_variance_ratio_)

# Variance cumulée
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
print(cumulative_variance)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR as SVM

model = SVM()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Prédiction sur l'ensemble de test
y_pred = model.predict(X_test)

# Calcul des métriques
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"R²: {r2}")
```

```python
# y_pred représente les prédictions de votre modèle
# y_true représente les valeurs réelles

diff = np.abs(y_pred - y_test)
tolerance = 15  # 15 minutes
within_tolerance = np.sum(diff <= tolerance)

percentage_within_tolerance = (within_tolerance / len(y_test)) * 100
print(f"Pourcentage de prédictions dans un seuil de tolérance de ±{tolerance} minutes: {percentage_within_tolerance:.2f}%")

```
